```python
import json
import pandas as pd
import numpy as np
```

```python
def load_data(file_name):
    lines = open(file_name).readlines()
    users_list = (json.loads(single_line) for single_line in lines)
    return pd.DataFrame(users_list)
```

```python
df = load_data('data/users.jsonl')
df
```

```python
df['business_address'][0]
```

```python
countries = []

for i in df.index:
    country = df['business_address'][i]['country']
    countries.append(country)
    
df['company_country'] = countries
```

```python
streets = []
for i in df.index:
    street = df['business_address'][i]['street']
    streets.append(street)
    
df['company_street'] = streets
```

```python
cities = []
for i in df.index:
    city = df['business_address'][i]['city']
    cities.append(city)
    
df['company_city'] = cities
```

```python
zip_codes = []
for i in df.index:
    code = df['business_address'][i]['zipcode']
    
    zip_codes.append(code)
    
df['company_zipcode'] = zip_codes
```

```python
df['full_name'] = df['first_name'] + ' ' +  df['last_name']
```

```python
df['primary_company'][0]
```

```python
titles = []
for i in df.index:
    job_title = df['primary_company'][i]['title']
    titles.append(job_title)
    
df['job_title'] = titles
```

```python
companies = []
for i in df.index:
    company = df['primary_company'][i]['name']
    companies.append(company)
    
df['company'] = companies
```

```python
numeric_gender = {"m": 0, "f": 1}
df['gender'] = df['gender'].map(numeric_gender)
```

```python
df = df.drop(['id'], axis=1)
```

```python
haves_set = set()

for haves_list in df['haves']:
    for have in haves_list:
        haves_set.add(have)

haves_vectors = []
for i in df.index:
    haves_vectors.append([1 if have in df['haves'][i] else 0 for have in haves_set])

df['haves_vector'] = haves_vectors
```

```python
def n_hot_encode(df, column):
    
    # get unique elements from df[column] and put them into a set
    _set = set()

    for _list in df[column]:
        for _elem in _list:
            _set.add(_elem)

    # for each row, compute it's encoded vector indicating with a 
    # 1 that it contains that element from the set and a 0 that it doesn't
    _vectors = []
    
    for i in df.index:
        _vectors.append([1 if _elem in df[column][i] else 0 for _elem in _set])

    return _vectors
```

```python
df['haves_vector'] = n_hot_encode(df, 'haves')
df['wants_vector'] = n_hot_encode(df, 'wants')
df['languages_vector'] = n_hot_encode(df, 'languages')
df
```
